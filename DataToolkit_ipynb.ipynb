Question:-4  Explain the use of the groupby() method in Pandas.

Answer: The groupby() method in Pandas is a powerful tool used to split a DataFrame into groups based on the values in one or more columns. After grouping, we can efficiently perform operations—such as calculations or transformations—on each group separately, and then combine the results back into a DataFrame or Series.




```python
import pandas as pd

data = {
    'car': ['Skoda', 'Skoda', 'Ford', 'Skoda', 'Ford', 'Ford'],
    'co2': [95, 90, 99, 104, 105, 94]
}
df = pd.DataFrame(data)
df.groupby('car').mean()
```

Question:6  What are the differences between NumPy arrays and Python lists?

Answer:The main differences between NumPy arrays and Python lists are centered on performance, data type handling, memory efficiency, and supported operations:-

Performance: NumPy arrays are much faster and more memory-efficient for numerical and scientific computing, especially when working with large datasets.

Homogeneity: NumPy arrays require all elements to be of the same data type, which enables vectorized operations and efficiency. Python lists can store elements of different types, which provides flexibility but reduces speed and efficiency.

Functionality: Many advanced mathematical functions (mean, sum, matrix multiplication) can be performed directly and efficiently on NumPy arrays, but require explicit loops or list comprehensions with Python lists.

Memory Layout: NumPy arrays store elements contiguously, leading to less memory fragmentation and better cache performance. Python lists store references, resulting in higher overhead.

Question:-8  What does the term “vectorized operation” mean in NumPy?

Answer:  Vectorized operation in NumPy refers to performing mathematical or logical operations directly on entire arrays, rather than using explicit loops to process each element one by one. This means that we can write code like array + 5 and have the value 5 added to every item in the array in a single, efficient step, rather than looping through each element and adding 5 individually.

Question:15. How does NumPy handle multidimensional arrays?

Answer: NumPy handles multidimensional arrays through its core object, the ndarray, which is designed to efficiently store and operate on arrays of arbitrary dimensions (also called axes or ranks).

NumPy manages and works with them:

1.Creation:  Multidimensional arrays can be created in several ways:By passing nested lists to np.array(), where each inner list represents a dimension (e.g., a list of lists forms a 2D array).Using functions like np.zeros(), np.ones(), or np.empty() with a shape tuple to specify the desired dimensions.

2.Shape and Axes:  Each array has a shape attribute describing the size of each dimension (e.g., (3, 4) for a 3-row, 4-column 2D array).
The number of axes the array has is called its rank (e.g., 2 for a 2D matrix, 3 for a 3D cube of values).

3.Indexing and Slicing: we can access or modify elements using multi-indexing, e.g., a[2] accesses the element in the first row and third column of a 2D array.
Slicing along any dimension is supported, enabling selection or modification of sub-arrays without copying data, which makes operations efficient.

4.Mathematical Operations: Mathematical operations apply element-wise across all dimensions (i.e., vectorized operations), regardless of array rank. For example, adding two arrays of the same shape or applying universal functions like np.sin() or np.sum() operate seamlessly on multidimensional arrays.
For true matrix multiplication (not just element-wise), functions like np.dot() or the @ operator are used for arrays with appropriate shapes.

5.Reshaping and Manipulation:
Arrays can be reshaped to different dimensions using reshape() provided the total number of elements remains the same (e.g., convert a flat array with 6 elements to a 2×3 matrix).

6.Memory Efficiency and Performance:
Arrays in NumPy are stored in contiguous memory blocks in C, allowing for high performance in numerical computations, even in multidimensional contexts.

map() Method
Scope: Works on Pandas Series (a single column of a DataFrame).

Purpose: Used for element-wise transformations.

Question:21. Why is Numpy's array slicing Faster than Python's list slicing?

Answer: NumPy's array slicing is faster than Python's list slicing primarily because:

1.Memory Efficiency through Views: NumPy slicing returns a view of the original array rather than creating a new copy, meaning it directly accesses the existing data buffer without additional memory allocation or copying. In contrast, Python list slicing creates a new list object, copying references to the sliced elements, which is more memory- and time-consuming.

2.Contiguous and Homogeneous Memory Storage: NumPy arrays store elements in a contiguous, fixed-type (homogeneous) block of memory. This layout allows fast, low-level pointer arithmetic to compute slice start points and strides efficiently. Python lists store elements as separate Python objects scattered in memory, so slicing involves more overhead.

3.Optimized C-Level Implementation: NumPy is implemented in C and uses optimized, low-level operations for indexing and slicing, bypassing the overhead of Python’s dynamic type checking and abstraction layers that lists require.

4.Support for Multidimensional Slicing: NumPy efficiently handles slicing in multiple dimensions with direct memory strides, enabling sophisticated slicing without slow Python overhead, which list slicing does not natively support.
